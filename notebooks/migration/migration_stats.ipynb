# Miami Migration and Population Exploratory Data Analysis

This notebook analyzes the inferred net migration flows for the state of Florida and Miami-Dade county based on historical population estimates.

```python
import os
import pandas as pd
import plotly.express as px

csv_path = '../../data/migration/migration_data.csv'
if os.path.exists(csv_path):
    df = pd.read_csv(csv_path, parse_dates=['Year'])
    print("Data loaded successfully.")
else:
    print(f"File not found: {csv_path}")
```

```python
# Project rules dictate Explicit handling of missing values (NaNs)
df.dropna(inplace=True)
df.head()
```

## Visualizing Florida Implied Net Migration

```python
fig1 = px.bar(df, x='Year', y='FL_Implied_Net_Migration_Thousands', title='Florida: Implied Net Migration (Thousands)')
fig1.show()
```

## Visualizing Miami-Dade County Implied Net Migration

```python
fig2 = px.bar(df, x='Year', y='Miami_Implied_Net_Migration_Thousands', title='Miami-Dade: Implied Net Migration (Thousands)')
fig2.show()
```
